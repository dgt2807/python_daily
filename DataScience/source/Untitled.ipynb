```python
raw_data=pd.read_excel('../data/seoul.xls',header=0,usecols="C:O")
raw_data=raw_data.iloc[3:9,:]
raw_data.set_index('분류',inplace=True)
raw_data= raw_data.drop('합계',axis=1)
```

```python
f, ax =plt.subplots(2,3,figsize=(16,8))

raw_data['휴식'].plot.pie(ax=ax[0][0],autopct='%1.1f%%')
ax[0][0].set_title('휴식')
ax[0][0].set_ylabel('')

raw_data['운동경기관람'].plot.pie(ax=ax[0][1],autopct='%1.1f%%')
ax[0][1].set_title('운동경기관람')
ax[0][1].set_ylabel('')

raw_data['창작적 취미활동'].plot.pie(ax=ax[0][2],autopct='%1.1f%%')
ax[0][2].set_title('창작적 취미활동')
ax[0][2].set_ylabel('')

raw_data['TV  또는 비디오시청'].plot.pie(ax=ax[1][0],autopct='%1.1f%%')
ax[1][0].set_title('TV  또는 비디오시청')
ax[1][0].set_ylabel('')

raw_data['여행야외나들이'].plot.pie(ax=ax[1][1],autopct='%1.1f%%')
ax[1][1].set_title('여행야외나들이')
ax[1][1].set_ylabel('')

raw_data['문화예술관람'].plot.pie(ax=ax[1][2],autopct='%1.1f%%')
ax[1][2].set_title('문화예술관람')
ax[1][2].set_ylabel('')
plt.show()
```
